# Crop Mapping Research Demo Dashboard

This dashboard showcases the results of the 3-stage spectro-temporal band selection method for crop mapping in Sacramento Valley, California.

```python
import os
import leafmap
import rasterio
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

# Constants from project config
KEEP_CLASSES = [3, 6, 24, 36, 37, 54, 69, 75, 76, 220, 61]
CLASS_NAMES = {
    0: "Background",
    3: "Rice", 6: "Sunflower", 24: "Winter Wheat",
    36: "Alfalfa", 37: "Other Hay", 54: "Tomatoes",
    69: "Grapes", 75: "Almonds", 76: "Walnuts",
    220: "Plums", 61: "Fallow"
}

CROP_COLORS = [
    "#000000", # 0: Background
    "#00a9e6", # 1: Rice (3)
    "#ffff00", # 2: Sunflower (6)
    "#a87000", # 3: Winter Wheat (24)
    "#ffa8e3", # 4: Alfalfa (36)
    "#a5f58d", # 5: Other Hay (37)
    "#f5a27a", # 6: Tomatoes (54)
    "#704489", # 7: Grapes (69)
    "#00a884", # 8: Almonds (75)
    "#ebd6b0", # 9: Walnuts (76)
    "#ff91ab", # 10: Plums (220)
    "#bfbf7a", # 11: Fallow (61)
]

LABEL_LIST = [CLASS_NAMES.get(c, str(c)) for c in [0, 3, 6, 24, 36, 37, 54, 69, 75, 76, 220, 61]]

# Data paths (relative to notebook in dashboard/ folder)
rgb_tif = "data/demo_s2_rgb.tif"
pred_tif = "data/demo_prediction_exp_C.tif"
gt_tif = "../data/processed/cdl/cdl_2022_study_area_filtered.tif"

print(f"RGB exists: {os.path.exists(rgb_tif)}")
print(f"Pred exists: {os.path.exists(pred_tif)}")
print(f"GT exists: {os.path.exists(gt_tif)}")
```

## Interactive Map Comparison

Use the map below to compare Sentinel-2 RGB imagery with the Ground Truth and Model Prediction.

```python
m = leafmap.Map(center=[39.15, -121.8], zoom=10)

if os.path.exists(rgb_tif):
    m.add_raster(rgb_tif, layer_name="Sentinel-2 RGB (2022-01-01)")

if os.path.exists(gt_tif):
    m.add_raster(gt_tif, palette=CROP_COLORS, layer_name="Ground Truth (CDL 2022)", opacity=0.7)

if os.path.exists(pred_tif):
    m.add_raster(pred_tif, palette=CROP_COLORS, layer_name="Prediction (Exp C - SegFormer)", opacity=0.7)

m.add_legend(title="Crop Classes", colors=CROP_COLORS, labels=LABEL_LIST)
m
```

## Dual-Map Split View

Swipe between Ground Truth and Prediction to see how well the model aligns with USDA CDL labels.

```python
m_split = leafmap.Map(center=[39.15, -121.8], zoom=12)
m_split.split_map(
    left_layer=gt_tif, 
    right_layer=pred_tif, 
    left_args={"palette": CROP_COLORS, "layer_name": "Ground Truth"}, 
    right_args={"palette": CROP_COLORS, "layer_name": "Prediction"}
)
m_split
```
