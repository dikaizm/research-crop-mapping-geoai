# 01 — Fetch Data
Paste any Google Drive link and download the file with its original name.

**Supported URL formats:**
- `https://drive.google.com/open?id=FILE_ID`
- `https://drive.google.com/file/d/FILE_ID/view`
- `https://drive.google.com/uc?id=FILE_ID`

```python
import gdown
import os
import re
import shutil


def parse_gdrive_id(url: str) -> str:
    """Extract file ID from any Google Drive URL format."""
    patterns = [
        r'[?&]id=([a-zA-Z0-9_-]+)',   # ?id=... or &id=...
        r'/file/d/([a-zA-Z0-9_-]+)',  # /file/d/FILE_ID/view
        r'/d/([a-zA-Z0-9_-]+)',       # short /d/FILE_ID
    ]
    for pattern in patterns:
        match = re.search(pattern, url)
        if match:
            return match.group(1)
    raise ValueError(f"Could not extract file ID from URL: {url}")


def fetch(url: str, output_dir: str, overwrite: bool = False) -> str:
    """
    Download a file from Google Drive using its shareable URL.
    The file is saved with its original filename from Google Drive.

    Parameters
    ----------
    url : str
        Any Google Drive shareable link.
    output_dir : str
        Directory to save the downloaded file.
    overwrite : bool
        If False and the file already exists, skip download.

    Returns
    -------
    str
        Full path to the downloaded file.
    """
    file_id = parse_gdrive_id(url)
    os.makedirs(output_dir, exist_ok=True)

    # Download to output_dir; gdown preserves original filename when output ends with /
    output_dir = output_dir.rstrip('/') + '/'
    output_path = gdown.download(id=file_id, output=output_dir, quiet=False, resume=not overwrite)

    if output_path is None:
        raise RuntimeError("Download failed. Check if the file is publicly shared.")

    print(f"\n✅ Saved to: {output_path}")
    return output_path
```
